```python
import pandas as pd 
import os

ROOT = "/Users/opolk/Documents/FAC/M1DS/DB/projet/ProjetBDDAvance"
os.chdir(ROOT)


sources_caracteristiques = {
    "caract_2005":   "data/raw/accidents_fr/caracteristiques_2005.csv",
    "caract_2007":   "data/raw/accidents_fr/caracteristiques_2007.csv",
    "caract_2009":   "data/raw/accidents_fr/caracteristiques_2009.csv",
    "caract_2011":   "data/raw/accidents_fr/caracteristiques_2011.csv",
    "caract_2013":   "data/raw/accidents_fr/caracteristiques_2013.csv",
    "caract_2015":   "data/raw/accidents_fr/caracteristiques_2015.csv",
    "caract_2017":   "data/raw/accidents_fr/caracteristiques_2017.csv",
    "caract_2019":   "data/raw/accidents_fr/caracteristiques_2019.csv",
    "caract_2021":   "data/raw/accidents_fr/caracteristiques_2021.csv",
}

sources_lieux = {
    "lieux_2005":   "data/raw/accidents_fr/lieux_2005.csv",
    "lieux_2007":   "data/raw/accidents_fr/lieux_2007.csv",
    "lieux_2009":   "data/raw/accidents_fr/lieux_2009.csv",
    "lieux_2011":   "data/raw/accidents_fr/lieux_2011.csv",
    "lieux_2013":   "data/raw/accidents_fr/lieux_2013.csv",
    "lieux_2015":   "data/raw/accidents_fr/lieux_2015.csv",
    "lieux_2017":   "data/raw/accidents_fr/lieux_2017.csv",
    "lieux_2019":   "data/raw/accidents_fr/lieux_2019.csv",
    "lieux_2021":   "data/raw/accidents_fr/lieux_2021.csv",
}

sources_usagers = {
    "usagers_2005":"data/raw/accidents_fr/usagers_2005.csv",
    "usagers_2007":"data/raw/accidents_fr/usagers_2007.csv",
    "usagers_2009":"data/raw/accidents_fr/usagers_2009.csv",
    "usagers_2011":"data/raw/accidents_fr/usagers_2011.csv",
    "usagers_2013":"data/raw/accidents_fr/usagers_2013.csv",
    "usagers_2015":"data/raw/accidents_fr/usagers_2015.csv",
    "usagers_2017":"data/raw/accidents_fr/usagers_2017.csv",
    "usagers_2019":"data/raw/accidents_fr/usagers_2019.csv",
    "usagers_2021":"data/raw/accidents_fr/usagers_2021.csv"
}

sources_vehicules = {
    "vehicules_2005":"data/raw/accidents_fr/vehicules_2005.csv",
    "vehicules_2007":"data/raw/accidents_fr/vehicules_2007.csv",
    "vehicules_2009":"data/raw/accidents_fr/vehicules_2009.csv",
    "vehicules_2011":"data/raw/accidents_fr/vehicules_2011.csv",
    "vehicules_2013":"data/raw/accidents_fr/vehicules_2013.csv",
    "vehicules_2015":"data/raw/accidents_fr/vehicules_2015.csv",
    "vehicules_2017":"data/raw/accidents_fr/vehicules_2017.csv",
    "vehicules_2019":"data/raw/accidents_fr/vehicules_2019.csv",
    "vehicules_2021":"data/raw/accidents_fr/vehicules_2021.csv"

}
```

### Concaténation des fichiers caractéristiques

```python
liste = []
for nom, chemin in sources_caracteristiques.items():
    if chemin == "data/raw/accidents_fr/caracteristiques_2009.csv": 
        separator = '\t'
    else:
        if chemin == "data/raw/accidents_fr/caracteristiques_2019.csv" or chemin == "data/raw/accidents_fr/caracteristiques_2021.csv":
            separator = ';'
        else:
            separator = ','

    df = pd.read_csv(chemin, sep=separator, encoding='latin-1')
    liste.append(df)

# pd.concat gère automatiquement le problème de l'ordre des colonnes mélangé, et la présence d'une colonne de plus dans les fichiers avant 2017
# => le fichier global contiendra une colonne 'gps' dont la valeur vaudra NaN pour les accidents de 2019 et 2021

caracteristiques = pd.concat(liste)

os.makedirs("data/combined", exist_ok=True)
caracteristiques.to_csv("data/combined/caracteristiques.csv", index=False)
```

### Concaténation des fichiers lieux


```python
liste.clear()
for nom, chemin in sources_lieux.items():
    if chemin == "data/raw/accidents_fr/lieux_2019.csv" or chemin == "data/raw/accidents_fr/lieux_2021.csv":
        separator = ';'
    else:
        separator = ','

    df = pd.read_csv(chemin, sep=separator, encoding='latin-1')
    liste.append(df)

lieux = pd.concat(liste)

lieux.to_csv("data/combined/lieux.csv", index=False)
```

### Concaténation des fichiers usagers

```python
liste.clear()
for nom, chemin in sources_usagers.items():
    if chemin == "data/raw/accidents_fr/usagers_2019.csv" or chemin == "data/raw/accidents_fr/usagers_2021.csv":
        separator = ';'
    else:
        separator = ','

    df = pd.read_csv(chemin, sep=separator, encoding='latin-1')
    liste.append(df)

usagers = pd.concat(liste)

usagers.to_csv("data/combined/usagers.csv", index=False)
```

### Concaténation des fichiers vehicules

```python
liste.clear()
for nom, chemin in sources_vehicules.items():
    if chemin == "data/raw/accidents_fr/vehicules_2019.csv" or chemin == "data/raw/accidents_fr/vehicules_2021.csv":
        separator = ';'
    else:
        separator = ','

    df = pd.read_csv(chemin, sep=separator, encoding='latin-1')
    liste.append(df)

vehicules = pd.concat(liste)

vehicules.to_csv("data/combined/vehicules.csv", index=False)
```

### Jointure des 4 fichiers combinés sur "Num_Acc"

```python
jointure = caracteristiques.merge(lieux, on="Num_Acc").merge(usagers, on="Num_Acc").merge(vehicules, on="Num_Acc")
print(jointure)
jointure.to_csv("data/combined/jointure.csv", index=False)
```
